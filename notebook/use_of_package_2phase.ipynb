This cell imports the py_graspi package as ig for easier reference throughout the code.

```python
#import py_graspi library
import py_graspi as ig
```

An example file is given and stored in the variable filename. This will be used in place of a command line argument since the functions need a .txt input file with proper formatting to generate a graph for the given file (Note: you can utilize your own properly formatted .txt file as long as the pathname of each file is correct)

```python
#Below is the pathname of an image to use for an example
filename = "../data/2phase/2D-testFiles/testFile-4x3-2D.txt"
```

The generateGraphAdj(filename) function below is from the py-graspi package is used to create a graph for any proper formated .txt file. It returns a graph_data_class where the newly generated graph is now stored as well as some other graph related information, these other variables are used when calculating descriptors for the file. (Note the inclusion of 'ig' before the function name which is needed to signal that you are reaching into the package for this given function).

```python
#This function will create a graph based on any given txt file input
graph_data = ig.generateGraphAdj(filename)
```

This cell uses the visualize function to display the graph for the data generated from the given morphology file. For explanation on what the colors of each vertex means, visit the py-graspi documentation.

```python
ig.visualize(graph_data.graph, graph_data.is_2D)
```

The following cell creates a dictionary holding the name and value of all descriptors generated by the compute_descriptors function from the py_graspi package. Then, we can use an example filename in the outputFile variable to be the output txt file. This can be adjusted as needed. Finally, the package function descriptorsToTxt writes the descriptors to the designated file path.

```python
#This function will create a tuple holding in the name and value of all descriptors
dict_descriptors = ig.compute_descriptors(graph_data, filename)

#We will use the filename below to be the output txt file
outputFile = "example_descriptors.txt"

#Copy descriptor tuple into txt file
ig.descriptorsToTxt(dict_descriptors, outputFile)
```

Finally to make sure the descriptors were properly written to the file we will now open the same outputFile in READ mode now and print out its contents and make sure we have the correct dictionary.

```python
#Output txt file, if error occurs this means txt file was not properly created
ig.readDescriptorsFromTxt(outputFile)
```
